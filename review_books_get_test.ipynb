```python
start = time.time()

url = "https://product.kyobobook.co.kr/detail/S000210412118"

# 크롬드라이버 실행
driver = webdriver.Chrome() 

# 화면 크기 설정
driver.set_window_size(1920, 1080)


#크롬 드라이버에 url 주소 넣고 실행
driver.get(url)

html = driver.page_source
soup = bs(html, 'html.parser')
```

```python
import requests
import json

bookUrl = 'https://product.kyobobook.co.kr/detail/S000210412118'

custom_header = {
    'referer' : f'{bookUrl}',
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36 Edg/119.0.0.0' 
}
dataUrl = 'https://product.kyobobook.co.kr/api/review/list?page=1&pageLimit=100000&reviewSort=001&revwPatrCode=000&saleCmdtid=S000210412118'

req = requests.get(dataUrl, headers=custom_header)

parsed_data = {}
if req.status_code == requests.codes.ok:    
    parsed_data = json.loads(req.text)
parsed_data

len(parsed_data['data']['reviewList'])

parsed_data['data']['reviewList'][0]['mmbrId']
parsed_data['data']['reviewList'][0]['cretDttm']
parsed_data['data']['reviewList'][0]['revwCntt']
parsed_data['data']['reviewList'][0]['revwRvgr']
```
